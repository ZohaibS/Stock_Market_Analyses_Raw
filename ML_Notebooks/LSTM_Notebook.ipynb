```python
#Splitting the Data into Training and Testing Sets 

raw_sets = data.values

training_set = raw_sets[:2404]
validation_set = raw_sets[2404:]
```

```python
np.shape(training_set)
```

```python
#Making our x trains and y trains

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(raw_sets)

x_train, y_train = [], []
for i in range(60,len(training_set)):
    x_train.append(scaled_data[i-60:i,0])
    y_train.append(scaled_data[i,0])
    print(np.shape(x_train))
x_train, y_train = np.array(x_train), np.array(y_train)

x_train = np.reshape(x_train, (x_train.shape[0],x_train.shape[1],1))


```

```python
#Forging a Long Short Term Memory Network

model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1],1)))
model.add(LSTM(units=50))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(x_train, y_train, epochs=1, batch_size=1, verbose=2)
```

```python
#Recursively Predicting 365 Days Worth of Data using the Previous 60

inputs = data[len(data) - len(validation_set) - 60:].values
inputs = inputs.reshape(-1,1)
inputs  = scaler.transform(inputs)

X_test = []
for i in range(len(data)-60,len(data)):
    X_test.append(inputs[i-60:i,0])
X_test = np.array(X_test)

# X_test = np.reshape(X_test, (X_test.shape[0],X_test.shape[1],1))
# closing_price = model.predict(X_test)
# closing_price = scaler.inverse_transform(closing_price)
# X_test
```

```python
X_predict = []
for i in range(60,inputs.shape[0]):
    X_predict.append(inputs[i-60:i,0])
    
X_predict = np.array(X_predict)

X_predict = np.reshape(X_predict, (X_predict.shape[0], X_predict.shape[1],1))
close_price = model.predict(X_predict)
close_price = scaler.inverse_transform(close_price)
X_predict
```

```python
#Our Error: In this Context its Root Mean Squared

root_mean_squared=np.sqrt(np.mean(np.power((validation_set-close_price),2)))
root_mean_squared
```
