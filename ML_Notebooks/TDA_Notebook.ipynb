```r
raw_csv <- read.csv(file="Signal_Data.csv", header=TRUE, sep=",")
Frequency <- raw_csv['Frequency']
Amplitude <- raw_csv['Amplitude']

Amp <- as.data.frame(Amplitude, colNames = 'Amplitude')
Freq <- as.data.frame(Frequency)
Freq_df <- data.frame(Frequency = Freq, Amplitude = Amplitude)
head(Freq_df)
```

```r
Y_1 <- Freq_df[1:405,]
Y_2 <- Freq_df[406:810,]
Y_3 <- Freq_df[811:1214,]
Y_4 <- Freq_df[1215:1619,]
Y_5 <- Freq_df[1620:2024,]
Y_6 <- Freq_df[2025:2429,]
Y_7 <- Freq_df[2330:2834,]
```
